```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人である。")
        else:
            print(f"{self.name}は大人ではない。")

humans = [
    Human("太郎", 20),
    Human("次郎", 19),
    Human("三郎", 30),
]

for human in humans:
    human.check_adult()
```
